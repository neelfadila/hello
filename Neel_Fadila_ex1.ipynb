1. Load breast cancer dataset (**structured data**)

For more details about the data: https://scikit-learn.org/1.5/modules/generated/sklearn.datasets.load_breast_cancer.html

```python


from sklearn.datasets import load_breast_cancer
my_data = load_breast_cancer()

```

2. Split **my_data** to train and test:

- Define X_train, X_test, Y_train, Y_test
- Choose **test_size** for splitting **my_data**
- Use **train_test_split** (for details: https://scikit-learn.org/dev/modules/generated/sklearn.model_selection.train_test_split.html)

```python


from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# טעינת הנתונים (גם אם לא רצו קודם)
my_data = load_breast_cancer()

X = my_data.data
Y = my_data.target

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42,
    stratify=Y
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)



```

3. Libraries

```python
!pip install mlflow
!pip install mlflow scikit-learn

import mlflow
import mlflow.sklearn
from mlflow import log_param, log_metric

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

import itertools
import pandas as pd
```

4. Define MLFlow experiment

```python
EXPERIMENT_NAME = "trees_hyperparam"
mlflow.set_experiment(EXPERIMENT_NAME)
# MLFlow details: https://mlflow.org/docs/latest/ml/tracking/
```

```python
results_rows = []
```

5. Train **model_decision_tree**

- Library: sklearn.tree.DecisionTreeClassifier
- Data: X_train, Y_train
- **Essential**: explore and optimize DecisionTreeClassifier options   

```python

from sklearn.tree import DecisionTreeClassifier

param_1_list = ["gini", "entropy"]      # criterion
param_2_list = [None, 3, 5, 7]          # max_depth
param_3_list = [2, 5, 10]               # min_samples_split

param_grid = list(itertools.product(param_1_list, param_2_list, param_3_list))

# אם עוד לא יצרת results_rows + calc_metrics, תוסיפי תא לפני זה (אני שולח אם צריך)

for param_1, param_2, param_3 in param_grid:
    with mlflow.start_run():
        mlflow.log_param("model_type", "DecisionTree")
        mlflow.log_param("criterion", param_1)
        mlflow.log_param("max_depth", param_2)
        mlflow.log_param("min_samples_split", param_3)

        d_tree = DecisionTreeClassifier(
            criterion=param_1,
            max_depth=param_2,
            min_samples_split=param_3,
            random_state=42
        )
        d_tree.fit(X_train, Y_train)

        y_pred = d_tree.predict(X_test)

        acc = accuracy_score(Y_test, y_pred)
        pre = precision_score(Y_test, y_pred, zero_division=0)
        rec = recall_score(Y_test, y_pred, zero_division=0)
        f1  = f1_score(Y_test, y_pred, zero_division=0)

        mlflow.log_metric("accuracy", acc)
        mlflow.log_metric("precision_score", pre)
        mlflow.log_metric("recall_score", rec)
        mlflow.log_metric("f1_score", f1)

        results_rows.append({
            "model_type": "DecisionTree",
            "criterion": param_1,
            "max_depth": param_2,
            "min_samples_split": param_3,
            "accuracy": acc,
            "precision_score": pre,
            "recall_score": rec,
            "f1_score": f1
        })
```

6. Train model_random_forest
- Library: sklearn.ensemble.RandomForestClassifier
- Data: X_train, Y_train
- **Essential**: explore and optimize RandomForestClassifier options

```python
from sklearn.ensemble import RandomForestClassifier

param_1_list = [50, 100, 200, 400]        # n_estimators
param_2_list = [None, 3, 5, 7, 10, 15]    # max_depth
param_3_list = ["sqrt", "log2"]           # max_features

param_grid = list(itertools.product(param_1_list, param_2_list, param_3_list))

for param_1, param_2, param_3 in param_grid:
    with mlflow.start_run():

        # Log parameters
        mlflow.log_param("model_type", "RandomForest")
        mlflow.log_param("n_estimators", param_1)
        mlflow.log_param("max_depth", param_2)
        mlflow.log_param("max_features", param_3)

        # Train the model
        rf = RandomForestClassifier(
            n_estimators=param_1,
            max_depth=param_2,
            max_features=param_3,
            random_state=42,
            n_jobs=-1
        )
        rf.fit(X_train, Y_train)

        # Predict + metrics
        y_pred = rf.predict(X_test)

        acc = accuracy_score(Y_test, y_pred)
        pre = precision_score(Y_test, y_pred, zero_division=0)
        rec = recall_score(Y_test, y_pred, zero_division=0)
        f1  = f1_score(Y_test, y_pred, zero_division=0)

        # Log metrics
        mlflow.log_metric("accuracy", acc)
        mlflow.log_metric("precision_score", pre)
        mlflow.log_metric("recall_score", rec)
        mlflow.log_metric("f1_score", f1)

        # Save results
        results_rows.append({
            "model_type": "RandomForest",
            "n_estimators": param_1,
            "max_depth": param_2,
            "max_features": param_3,
            "accuracy": acc,
            "precision_score": pre,
            "recall_score": rec,
            "f1_score": f1
        })
```

7. Train model_adaboost

- Library: sklearn.ensemble.AdaBoostClassifier
- Data: X_train, Y_train
- **Essential**: explore and optimize AdaBoostClassifier options

```python
from sklearn.ensemble import AdaBoostClassifier

param_1_list = [50, 100, 200, 400, 600, 800]   # n_estimators
param_2_list = [0.01, 0.05, 0.1, 0.2, 0.5]     # learning_rate
param_3_list = ["SAMME"]                       # algorithm

param_grid = list(itertools.product(param_1_list, param_2_list, param_3_list))

for param_1, param_2, param_3 in param_grid:
    with mlflow.start_run():

        # Log parameters
        mlflow.log_param("model_type", "AdaBoost")   # תיקון! לא RandomForest
        mlflow.log_param("n_estimators", param_1)
        mlflow.log_param("learning_rate", param_2)
        mlflow.log_param("algorithm", param_3)

        ada = AdaBoostClassifier(
            n_estimators=param_1,
            learning_rate=param_2,
            algorithm=param_3,
            random_state=42
        )
        ada.fit(X_train, Y_train)

        y_pred = ada.predict(X_test)

        acc = accuracy_score(Y_test, y_pred)
        pre = precision_score(Y_test, y_pred, zero_division=0)
        rec = recall_score(Y_test, y_pred, zero_division=0)
        f1  = f1_score(Y_test, y_pred, zero_division=0)

        # Log metrics
        mlflow.log_metric("accuracy", acc)
        mlflow.log_metric("precision_score", pre)
        mlflow.log_metric("recall_score", rec)
        mlflow.log_metric("f1_score", f1)

        results_rows.append({
            "model_type": "AdaBoost",
            "n_estimators": param_1,
            "learning_rate": param_2,
            "algorithm": param_3,
            "accuracy": acc,
            "precision_score": pre,
            "recall_score": rec,
            "f1_score": f1
        })
```

8. Store the result

```python
from google.colab import files

# יצירת טבלה מכל התוצאות
df = pd.DataFrame(results_rows)

# ניקוי עמודות זמן (אם קיימות)
df = df.drop(columns=[col for col in df.columns if "time" in col.lower()], errors="ignore")

print("Total runs saved:", len(df))
display(df.head())

# הצגת הטוב ביותר לכל מודל לפי F1
best_per_model = df.sort_values("f1_score", ascending=False).groupby("model_type").head(1)
print("Best run per model (by f1_score):")
display(best_per_model[["model_type", "accuracy", "precision_score", "recall_score", "f1_score"]])

# שמירה לאקסל
out_name = "student_name_results.xlsx"   # תשני לשם שלך/שלכם אם רוצים
df.to_excel(out_name, index=False)

# הורדה למחשב
files.download(out_name)
```
